# Fit SINDy model to the natural dynamics

```python
data = np.loadtxt('../transient/output/force.dat')
t, L, D = data.T
```

```python
plt.plot(t, L)
plt.grid()
```

```python
dt = t[1]-t[0]

def fd_deriv(x, dt=dt):
    dx = np.zeros_like(x)
    dx[1:-1] = (x[2:] - x[:-2])/(2*dt)
    dx[0] = (x[1]-x[0])/dt
    dx[-1] = (x[-1]-x[-2])/dt
    return dx
```

```python
dL = fd_deriv(L)
plt.plot(L, dL)
```

```python
x_train = np.vstack([L, dL]).T

# Instantiate and fit the SINDy model
lib = ps.PolynomialLibrary(degree=3)

# opt = ps.SR3(threshold=0.1, thresholder='l0')
opt = ps.STLSQ(threshold=1e-2)
model = ps.SINDy(feature_library=lib, optimizer=opt)
model.fit(x_train, t=dt)
model.print()
```

```python
t_test = t
x_test = x_train
x0_test = x_test[0, :]

dx_pred = model.predict(x_test)
dx_fd = model.differentiate(x_test, t=dt)

fig, ax = plt.subplots(1, 1, figsize=(12, 2))
ax.plot(t_test, dx_pred[:, 1], 'k', label='model prediction')
ax.plot(t_test, dx_fd[:, 1], 'r--', label='finite difference')
ax.set(xlabel='t', ylabel='$C_L$')

fig.show()
```

```python
x0_test
```

```python
t_test = t
x_test = x_train
x0_test = np.array([-1e-5, 0.0])
x_test_sim = model.simulate(x0_test, t_test)

fig, ax = plt.subplots(1, 1, figsize=(8, 2))
ax.plot(t_test, x_test[:, 0], 'k', label='DNS')
ax.plot(t_test, x_test_sim[:, 0], c='xkcd:red orange', label='SINDy')
ax.set(xlabel='t', ylabel='$C_L$')

ax.grid()
fig.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
ax.plot(x_test[:, 0], x_test[:, 1], c='k', label='DNS')
ax.plot(x_test_sim[:, 0], x_test_sim[:, 1], c='xkcd:red orange', label='SINDy')

ax.grid()
fig.show()
```

# Add forcing term 

```python
data = np.loadtxt('noise-response/force.dat')
t, L, D, u = data[1:, :].T
dL = fd_deriv(L)

x_ctrl = np.vstack([L, dL]).T
dx_pred = model.predict(x_ctrl)[2:-2, :]
dx_fd = model.differentiate(x_ctrl, t=dt)[2:-2, :]
t = t[2:-2]
u = u[2:-2]
L = L[2:-2]
D = D[2:-2]
```

```python
fig, axs = plt.subplots(2, 1, sharex=True, figsize=(8, 4))
axs[0].plot(t, L, 'k', label='DNS')
axs[0].grid()
axs[1].set(ylabel='$C_L$')

axs[1].plot(t, u, 'k', label='DNS')
axs[1].grid()
axs[1].set(xlabel='t', ylabel='$u$')

fig.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 2))
ax.plot(t_ctrl, dx_pred[:, 1] - dx_fd[:, 1], 'k', label='model discrepancy')
ax.set(xlabel='t', ylabel=r'$\ddot{C}_L$, $u$')

ax.plot(t_ctrl, u, 'xkcd:brick red', label='control')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 2))
ax.plot(t_ctrl, dx_pred[:, 0] - dx_fd[:, 0], 'k', label='model discrepancy')
ax.plot(t_ctrl, u, 'xkcd:brick red', label='control')
ax.set(xlabel='t', ylabel=r'$\dot{C}_L$, $u$')
```

```python
# Linear fit for B*u
delta = dx_pred[:, 1] - dx_fd[:, 1]
B = np.dot(delta, u)/np.dot(u, u)
print(B)

fig, ax = plt.subplots(1, 1, figsize=(12, 2))
ax.plot(t_ctrl, delta, 'k', label='model discrepancy')
ax.plot(t_ctrl, B*u, 'xkcd:red orange', label='control prediction')
ax.set(xlabel='t', ylabel=r'$\ddot{C}_L$')
```
